# **Disgust**

```python
path_kaggle1_disgust = "../data/raw/muhammadummarattique_data/disgust"
path_kaggle2_disgust = "../data/raw/emotions_description_RadeDacevic.xlsx"

path_semeval_dev = "../data/raw/semeval/2018-E-c-En-dev.txt"
path_semeval_test = "../data/raw/semeval/2018-E-c-En-test.txt"
path_semeval_train = "../data/raw/semeval/2018-E-c-En-train.txt"

path_disgust = "../data/raw/disgust_original.csv"
```

## Disgust 1 ([Kaggle link](https://www.kaggle.com/datasets/muhammadumarattique/text-for-emotions-classification-projests))

```python
disgust_texts = []

for filename in os.listdir(path_kaggle1_disgust):
    if filename.endswith('.txt'):
        filepath = os.path.join(path_kaggle1_disgust, filename)
        with open(filepath, 'r', encoding='utf-8') as f:
            content = f.read().strip()
            disgust_texts.append({'text': content, 'label': 'disgust'})

df1_d = pd.DataFrame(disgust_texts)
df1_d.head()
```

```python
df1_d.shape
```

## Semeval-2018 Task E-c ([Github link](https://github.com/cbaziotis/ntua-slp-semeval2018/tree/master/datasets/task1/E-c))

```python
import pandas as pd

base_raw_url = "https://raw.githubusercontent.com/cbaziotis/ntua-slp-semeval2018/master/datasets/task1/E-c/"

files = {
    "train": "E-c-En-train.txt",
    "dev": "E-c-En-dev.txt",
    "test": "E-c-En-test-gold.txt"
}

dataframes = []
for split, filename in files.items():
    url = base_raw_url + filename
    df = pd.read_csv(url, sep='\t', header=0, encoding='utf-8')
    df['split'] = split
    dataframes.append(df)

df3_d = pd.concat(dataframes, ignore_index=True)
df3_d.head()
```

```python
df3_d.dtypes
```

```python
df3_d = df3_d.rename(columns={"Tweet": "text"})
df3_d = df3_d[df3_d["disgust"]==1]
df3_d.shape
```

```python
df3_d = df3_d[df3_d["disgust"].astype(str).eq("1")].copy()
df3_d["label"] = "disgust"
df3_d = df3_d[["text", "label"]]
df3_d.drop_duplicates(subset=['text'], inplace=True)
df3_d.shape
```

## Merge disgust datasets

```python
df_merged_d = pd.concat([df1_d, df2_d, df3_d], ignore_index=True)
df_merged_d.head()
```

```python
df_merged_d.shape[0]
```

```python
df_merged_d.drop_duplicates(subset=['text'], inplace=True)
df_merged_d.shape[0]
```

```python
df_merged_d.to_csv(path_disgust, index=False)
print("Saved merged disgust dataset -> ", path_disgust)
```
